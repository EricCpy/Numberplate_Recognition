```python
import numpy as np
import matplotlib.pyplot as plt

thresholds = np.linspace(0.0, 1.0, 101)

def plot_metric_vs_confidence(thresholds, metric_values, labels, ylabel, title):
    plt.figure(figsize=(8, 6))
    for metric, label in zip(metric_values, labels):
        plt.plot(thresholds, metric, label=label)
    plt.xlabel('Confidence Threshold')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_recall_vs_precision(recalls, precisions, labels):
    plt.figure(figsize=(8, 6))
    for recall, precision, label in zip(recalls, precisions, labels):
        plt.plot(recall, precision, label=label)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision vs Recall')
    plt.legend()
    plt.grid(True)
    plt.show()
```
